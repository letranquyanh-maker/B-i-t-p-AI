```python
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
import folium


m = folium.Map(location=[10.761355552352372, 106.6683459252726], zoom_start=12)


folium.Marker(
    [10.761355552352372, 106.6683459252726], popup='UEH', tooltip='click').add_to(m)


geolocator = Nominatim(user_agent='my_map')


ds = [
    'Ben Thanh Market, Ho Chi Minh City, Vietnam',
    'Notre Dame Cathedral Basilica of Saigon, Ho Chi Minh City, Vietnam',
    'Independence Palace, Ho Chi Minh City, Vietnam',
    'Tan Son Nhat Airport, Ho Chi Minh City, Vietnam',
    'Landmark 81, Ho Chi Minh City, Vietnam'
]


for i in ds:
    x = geolocator.geocode(i)
    if x:
        diem = [x.latitude, x.longitude]
        kc = geodesic([10.761355552352372, 106.6683459252726], diem).km


        folium.Marker(diem, popup=i, tooltip='click').add_to(m)


        folium.PolyLine([[10.761355552352372, 106.6683459252726], diem], popup=str(round(kc, 2)) + ' km').add_to(m)

m
```
