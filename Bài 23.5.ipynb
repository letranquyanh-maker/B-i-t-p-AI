```python
import folium
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

trung_tam = [10.775134342570790, 106.690618301257354]

khach_hang = [
    [10.761208381640975, 106.66835646578188],
    [10.783304116512499, 106.69462528061219],
    [10.773376523822355, 106.67750683643312],
    [10.790579315471804, 106.69939015362725],
    [10.801465832740104, 106.71003110129301],
    [10.798830127838270, 106.68241198490234],
    [10.812341038020885, 106.67712490140891],
    [10.823095094142098, 106.62961901401062],
    [10.754709203403092, 106.66328945982438],
    [10.741209184011519, 106.70709185098512],
    [10.729109813419184, 106.79218491121593],
    [10.841161001109316, 106.80989123118733]
]

def tinh_khoang_cach(lat1, lon1, lat2, lon2):
    R = 6371000

    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)

    a = sin(dlat / 2) ** 2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c

def xac_dinh_vung(khoang_cach):
    if khoang_cach <= 3000:
        return 'Trong ban kinh 3 km'
    elif khoang_cach <= 5000:
        return 'Trong ban kinh 5 km'
    elif khoang_cach <= 10000:
        return 'Trong ban kinh 10 km'
    else:
        return 'Ngoai vung phuc vu'

ket_qua = []

for i in range(len(khach_hang)):
    vi_do = khach_hang[i][0]
    kinh_do = khach_hang[i][1]

    khoang_cach = tinh_khoang_cach(trung_tam[0], trung_tam[1], vi_do, kinh_do)
    vung = xac_dinh_vung(khoang_cach)

    ket_qua.append([i + 1, vi_do, kinh_do, khoang_cach, vung])

df = pd.DataFrame(ket_qua, columns=['khach', 'vi_do', 'kinh_do', 'khoang_cach', 'vung_phuc_vu'])

print(df)

m = folium.Map(location=trung_tam, zoom_start=11)

folium.Marker(
    trung_tam,
    popup='Trung tam phan phoi',
    tooltip='click',
    icon=folium.Icon(color='black', icon='home')
).add_to(m)

ban_kinh = [3000, 5000, 10000]
mau_vung = ['green', 'orange', 'red']

for i in range(len(ban_kinh)):
    folium.Circle(location=trung_tam,
        radius=ban_kinh[i],
        color=mau_vung[i],
        fill=True,
        fill_opacity=0.08,
        popup='Vung phuc vu ban kinh ' + str(int(ban_kinh[i] / 1000)) + ' km'
    ).add_to(m)

for i in range(len(df)):
    folium.CircleMarker(
        location=[df.loc[i, 'vi_do'], df.loc[i, 'kinh_do']],
        radius=6,
        color='blue',
        fill=True,
        fill_color='blue',
        popup='Khach ' + str(df.loc[i, 'khach']) + ' - ' + df.loc[i, 'vung_phuc_vu']
    ).add_to(m)

m
```
