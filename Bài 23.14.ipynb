```python
import osmnx as ox
import networkx as nx
import folium
from IPython.display import display, clear_output
import time

tam_ban_do = (10.7767, 106.7031)

G = ox.graph_from_point(
    tam_ban_do,
    dist=9000,
    network_type="drive"
)

diem_dau = (10.7767, 106.7031)
diem_cuoi = (10.8016, 106.6520)

def tim_nut_gan_nhat(toa_do):
    vi_do, kinh_do = toa_do
    return ox.distance.nearest_nodes(G, kinh_do, vi_do)

nut_dau = tim_nut_gan_nhat(diem_dau)
nut_cuoi = tim_nut_gan_nhat(diem_cuoi)

duong_di = nx.shortest_path(
    G,
    nut_dau,
    nut_cuoi,
    weight="length"
)

toa_do_duong = []

for nut in duong_di:
    vi_do = G.nodes[nut]["y"]
    kinh_do = G.nodes[nut]["x"]
    toa_do_duong.append((vi_do, kinh_do))

for t in range(len(toa_do_duong)):
    vi_tri_hien_tai = toa_do_duong[t]

    m = folium.Map(location=vi_tri_hien_tai, zoom_start=14)

    folium.PolyLine(
        toa_do_duong,
        color="blue",
        weight=4,
        opacity=0.6,
        popup="Tuyến đường di chuyển"
    ).add_to(m)

    folium.Marker(
        diem_dau,
        popup="Điểm xuất phát",
        icon=folium.Icon(color="green")
    ).add_to(m)

    folium.Marker(
        diem_cuoi,
        popup="Điểm kết thúc",
        icon=folium.Icon(color="red")
    ).add_to(m)

    folium.Marker(
        vi_tri_hien_tai,
        popup="Xe đang ở bước thời gian " + str(t),
        icon=folium.Icon(color="blue", icon="car", prefix="fa")
    ).add_to(m)

    clear_output(wait=True)
    print("Bước thời gian:", t)
    display(m)
    time.sleep(0.1)
```
