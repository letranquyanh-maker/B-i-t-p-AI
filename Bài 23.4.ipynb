```python
import folium
import pandas as pd

du_lieu = [
    ['Quan 1', 10.77696234, 106.70091234, 210000],
    ['Quan 3', 10.78431460, 106.68425750, 190000],
    ['Quan 5', 10.75425620, 106.61351260, 160000],
    ['Quan 7', 10.73574580, 106.72133450, 320000],
    ['Quan 10', 10.71345730, 106.65684670, 240000],
    ['Binh Thanh', 10.23468050, 10684567100, 490000],
    ['Go Vap', 10.82346380, 106.12346650, 680000],
    ['Tan Binh', 10.80567810, 106.62456520, 470000],
    ['Phu Nhuan', 10.71346980, 106.64794810, 180000],
    ['Thu Duc', 10.84123710, 106.82346090, 620000]
]

df = pd.DataFrame(du_lieu, columns=['ten_quan', 'vi_do', 'kinh_do', 'dan_so'])

def tao_vung(vi_do, kinh_do, do_rong):
    vung = [
        [kinh_do - do_rong, vi_do - do_rong],
        [kinh_do + do_rong, vi_do - do_rong],
        [kinh_do + do_rong, vi_do + do_rong],
        [kinh_do - do_rong, vi_do + do_rong],
        [kinh_do - do_rong, vi_do - do_rong]
    ]

    return vung

geojson = {
    'type': 'FeatureCollection',
    'features': []
}

for i in range(len(df)):
    ten_quan = df.loc[i, 'ten_quan']
    vi_do = df.loc[i, 'vi_do']
    kinh_do = df.loc[i, 'kinh_do']

    vung_quan = tao_vung(vi_do, kinh_do, 0.012)

    feature = {
        'type': 'Feature',
        'properties': {
            'ten_quan': ten_quan
        },
        'geometry': {
            'type': 'Polygon',
            'coordinates': [vung_quan]
        }
    }

    geojson['features'].append(feature)

m = folium.Map(location=[10.780, 106.700], zoom_start=11)

folium.Choropleth(
    geo_data=geojson,
    data=df,
    columns=['ten_quan', 'dan_so'],
    key_on='feature.properties.ten_quan',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.4,
    legend_name='Dan so theo quan'
).add_to(m)

folium.GeoJson(
    geojson,
    tooltip=folium.GeoJsonTooltip(
        fields=['ten_quan'],
        aliases=['Ten quan:']
    )
).add_to(m)

for i in range(len(df)):
    folium.Marker(
        [df.loc[i, 'vi_do'], df.loc[i, 'kinh_do']],
        popup=df.loc[i, 'ten_quan'] + ' - Dan so: ' + str(df.loc[i, 'dan_so']),
        tooltip='click'
    ).add_to(m)

m
```
