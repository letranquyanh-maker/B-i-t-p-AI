```python
import numpy as np
import pandas as pd
import folium
from sklearn.cluster import KMeans

np.random.seed(42)

tam_ban_do = [10.7767, 106.7031]
n = 120

vi_do = np.random.uniform(10.70, 10.86, n)
kinh_do = np.random.uniform(106.60, 106.77, n)

df = pd.DataFrame({
    "vi_do": vi_do,
    "kinh_do": kinh_do
})

so_tram = 4

X = df[["vi_do", "kinh_do"]]

model = KMeans(
    n_clusters=so_tram,
    random_state=42,
    n_init=10
)

df["cum"] = model.fit_predict(X)

tam_cum = model.cluster_centers_

m = folium.Map(location=tam_ban_do, zoom_start=11)

mau_cum = ["red", "blue", "green", "purple"]

for i in range(len(df)):
    cum = int(df.loc[i, "cum"])
    mau = mau_cum[cum]

    folium.CircleMarker(
        location=[float(df.loc[i, "vi_do"]), float(df.loc[i, "kinh_do"])],
        radius=4,
        color=mau,
        fill=True,
        fill_color=mau,
        fill_opacity=0.7,
        popup="Khách hàng thuộc cụm " + str(cum + 1),
        tooltip="Khách hàng"
    ).add_to(m)

for i in range(len(df)):
    cum = int(df.loc[i, "cum"])

    folium.PolyLine(
        locations=[
            [float(df.loc[i, "vi_do"]), float(df.loc[i, "kinh_do"])],
            [float(tam_cum[cum][0]), float(tam_cum[cum][1])]
        ],
        color=mau_cum[cum],
        weight=1,
        opacity=0.25
    ).add_to(m)

for i in range(len(tam_cum)):
    folium.Marker(
        location=[float(tam_cum[i][0]), float(tam_cum[i][1])],
        popup="Điểm trung chuyển đề xuất " + str(i + 1),
        tooltip="Trạm đề xuất " + str(i + 1),
        icon=folium.Icon(color="black", icon="home")
    ).add_to(m)

m
```
