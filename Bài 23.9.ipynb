```python
import folium
import pandas as pd
from sklearn.cluster import KMeans

khach_hang = [
    [10.761208381640975, 106.66835646578188],
    [10.783304116512499, 106.69462528061219],
    [10.773376523822357, 106.67750683643382],
    [10.790579315471804, 106.69939015362725],
    [10.801465563456637, 106.71003125724577],
    [10.798830835636736, 106.68241436735673],
    [10.812345545254277, 106.67789175247246],
    [10.823095245725472, 106.62966245725472],
    [10.754792562457272, 106.66343825625642],
    [10.741219123441234, 106.70751296780695],
    [10.729104346134613, 106.72159389557805],
    [10.841166365836832, 106.80989342897394]
] # dữ liệu tọa độ 12 khách hàng

df = pd.DataFrame(khach_hang, columns=['vi_do', 'kinh_do']) # chuyển dữ liệu khách hàng thành bảng để dễ xử lí

kmeans = KMeans(n_clusters=3, random_state=0, n_init=10) # tạo mô hình KMeans, chia khách hàng thành 3 cụm
df['cum'] = kmeans.fit_predict(df[['vi_do', 'kinh_do']]) # phân cụm dựa trên vĩ độ và kinh độ của khách hàng

tam_cum = kmeans.cluster_centers_ # lấy tọa độ tâm của 3 cụm, đây là vị trí đề xuất đặt trạm xe lun

print(df)
print('Tam cac cum:')
print(tam_cum)

m = folium.Map(location=[10.780, 106.700], zoom_start=11) # tạo bản đồ ở khu vực TP.HCM

mau = ['red', 'blue', 'green']

for i in range(len(df)):
    folium.CircleMarker(
        location=[df.loc[i, 'vi_do'], df.loc[i, 'kinh_do']], # lấy tọa độ của từng khách hàng trong bảng
        radius=7,
        color=mau[df.loc[i, 'cum']],
        fill=True,
        fill_color=mau[df.loc[i, 'cum']],
        popup='Khach ' + str(i + 1) + ' - Cum ' + str(df.loc[i, 'cum'])
    ).add_to(m)

for i in range(len(tam_cum)):
    folium.Marker(
        [tam_cum[i][0], tam_cum[i][1]], # tọa độ tâm cụm, xem như vị trí đề xuất đặt trạm xe
        popup='Vi tri de xuat dat tram xe cho cum ' + str(i), tooltip = 'click').add_to(m)

for i in range(len(df)):
    cum = df.loc[i, 'cum']

    folium.PolyLine(
        [[df.loc[i, 'vi_do'], df.loc[i, 'kinh_do']], [tam_cum[cum][0], tam_cum[cum][1]]], color=mau[cum], weight=2,
        popup='Khach ' + str(i + 1) + ' den tram cua cum ' + str(cum)).add_to(m) # nối khách hàng với trạm để nhìn cụm rõ hơn

m

```
